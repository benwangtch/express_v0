```python
import pandas as pd
import math
import numpy as np
```

```python
test= pd.read_csv('./final_inference/lgbm_透天厝.csv')
```

```python
# input = [303893.418, 2773228.853]
input = [169784.98804265473,2543493.999829864]
addr = '台南市東區青年路454號'
house_age = 33
far = 1.58
# '建物移轉總面積(坪)' : 3.484  '土地移轉總面積(坪)': 3.024 label :331437.79557039525 pred:307335.56736823893
```

```python
compare = []
for idx,item in test.iterrows():
    tmpLoc = [float(item['x座標']),float(item['y座標'])]
    tmp = math.dist(tmpLoc, input)
    compare.append(tmp)
```

```python
compare = np.array(compare)
idx = np.argpartition(compare,20)
print(idx)
```

```python
output = test.iloc[idx[:20]]
```

```python
compare_age = []
for idx,item in output.iterrows():
    tmpage = item['house_age']
    tmp = abs(tmpage - 33)
    compare_age.append(tmp)
```

```python
compare_age = np.array(compare_age)
idx_age = np.argpartition(compare_age,10)
print(idx_age)
```

```python
output_age = output.iloc[idx_age[:10]]
```
